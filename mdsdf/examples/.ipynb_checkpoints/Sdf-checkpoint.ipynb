```python
from mrsdf import Sdf
from graphviz import Source
```

```python
sdf = Sdf(['A', 'B', 'C'])
sdf.add_channel(source='A', target='B', production_rate=2, consumption_rate=3, initial_tokens=0)
sdf.add_channel(source='B', target='C', production_rate=1, consumption_rate=2, initial_tokens=0)
```

```python
Source(sdf.dot())
```

```python
hsdf = sdf.hsdf()
```

```python
Source(hsdf.dot())
```
