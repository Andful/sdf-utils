```python
from cyclic_scheduler import CyclicScheduler
```

```python
cs = CyclicScheduler()
```

```python
cs.add_task
```

```python
t1 = cs.add_task("a", 1, 0, 'red');
t2 = cs.add_task("b", 1, 1, 'green');
t3 = cs.add_task("c", 1, 0, 'blue');
```

```python
memory = cs.add_memory(100, 100)
```

```python
ring_buffer1 = cs.add_ring_buffer((1, 0), memory)
ring_buffer2 = cs.add_ring_buffer((1, 0), memory)
```

```python
cs.add_dependency(t1, t2, (2, 1), (3, 1), ring_buffer=ring_buffer1)
cs.add_dependency(t2, t3, (1, 1), (2, 1), ring_buffer=ring_buffer2)
```

```python
solution = cs.solve()
```

```python
solution.plot()
```
